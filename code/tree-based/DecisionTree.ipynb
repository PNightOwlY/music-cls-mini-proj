```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
song = pd.read_csv('../../data/training_data.csv', na_values='?').dropna()

song.sample(5)
```

```python
attr = ['acousticness','danceability','duration',
        'energy','instrumentalness','key','liveness',
        'loudness','mode','speechiness','tempo',
        'time_signature','valence']
label = 'label'

# the colinearity between different feature by geometric
# pd.plotting.scatter_matrix(song[attr], figsize = (10,10))
# plt.show()

# the colinearity between different feature by numeric
# song[attr].corr()
```

## Decision Tree


```python
X_train, X_test, Y_train, Y_test = train_test_split(song[attr],
                                                   song[label],
                                                   test_size=0.2,
                                                   random_state=1)

decision_tree = DecisionTreeClassifier(random_state=1, splitter='best')

param = {'criterion':['entropy','gini'],
         'max_depth':[5,6,7,8],
         'ccp_alpha':np.linspace(0,0.05,20)}

grid_searcher = GridSearchCV(decision_tree, param, cv=10)


grid_searcher.fit(X_train, Y_train)

print("Best model parameters \n{}".format(grid_searcher.best_params_))
print("best score:{}".format(grid_searcher.best_score_))

y_pred = grid_searcher.predict(X_test)
score = accuracy_score(Y_test, y_pred)

print(score)
```

```python
decision_tree = DecisionTreeClassifier(ccp_alpha=grid_searcher.best_params_['ccp_alpha'],
                                       max_depth=grid_searcher.best_params_['max_depth'],
                                       criterion=grid_searcher.best_params_['criterion'],
                                       random_state=1,
                                       splitter='best')

decision_tree.fit(X_train, Y_train)

y_pred = decision_tree.predict(X_test)
score = accuracy_score(Y_test, y_pred)

print(score)
```

```python
from IPython.display import Image  
import pydotplus

dot_data = export_graphviz(decision_tree, out_file=None, 
                         feature_names = attr, 
                         class_names = np.array(['dislike','like']),
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())
```

## Consider we want to see the entropy graph

```python
decision_tree = DecisionTreeClassifier(ccp_alpha=grid_searcher.best_params_['ccp_alpha'],
                                       max_depth=grid_searcher.best_params_['max_depth'],
                                       criterion='entropy',
                                       random_state=1,
                                       splitter='best')

decision_tree.fit(X_train, Y_train)

y_pred = decision_tree.predict(X_test)
score = accuracy_score(Y_test, y_pred)

print(score)
```

```python
dot_data = export_graphviz(decision_tree, out_file=None, 
                         feature_names = attr, 
                         class_names = np.array(['dislike','like']),
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())
```

**observation:** The clear difference between the CART and IDL tree is the branch, the branch of the ID3 is much wider, since the information gain criterion has a preference for features with a larger number of possible values. CART is a binary tree with low computation cost, but with small training, data is more likely to underfit.
